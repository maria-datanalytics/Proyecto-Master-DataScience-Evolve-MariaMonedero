# Raw → Bronze

```python
# ============================================================
# RAW → BRONZE
# Enterprise Finance Medallion Architecture
# ============================================================

from pyspark.sql.functions import current_timestamp, lit

# ============================================================
# READ RAW FILE
# ============================================================

raw_path = "abfss://MariaMonedero@onelake.dfs.fabric.microsoft.com/lh_enterprise_raw.Lakehouse/Files/ORIONTECH_raw.csv"

df_bronze = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "false")
    .csv(raw_path)
)

# ============================================================
# INGESTION METADATA
# ============================================================

df_bronze = (
    df_bronze
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_file", lit("ORIONTECH_raw.csv"))
    .withColumn("source_layer", lit("raw"))
)

# ============================================================
# VALIDATION
# ============================================================

rows_loaded = df_bronze.count()
columns_loaded = len(df_bronze.columns)

print("========================================")
print("RAW DATA INGESTION COMPLETED")
print("========================================")
print(f"Rows imported successfully: {rows_loaded}")
print(f"Columns detected: {columns_loaded}")

display(df_bronze)

# ============================================================
# SAVE BRONZE TABLE
# ============================================================

df_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze_fact_operations_raw")

print("========================================")
print("BRONZE TABLE CREATED SUCCESSFULLY")
print("========================================")
print("Table name: bronze_fact_operations_raw")
```
